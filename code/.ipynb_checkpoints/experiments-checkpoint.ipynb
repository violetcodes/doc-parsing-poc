### Open docs and align them them against target

```python
def get_files_path(folder_path): # subfolder is either 
    '''get function for getting list of file paths'''
    for i in os.listdir(folder_path):
        path_to_return = os.path.join(folder_path, i)
        if os.path.isfile(path_to_return):
            yield path_to_return

training_files_path_generator = get_files_path(TRAINING_FOLDER)
training_files_path_list = [i for i in get_files_path(TRAINING_FOLDER)]
```

```python
def get_full_text(file_path):
    document = docx.Document(file_path)
    return '\n\n'.join([para.text for para in document.paragraphs])
```

```python
def get_candidates(text):
    candidates_for_date = extract.extract_date_spans(text)
    candidates_for_date =[[text[start:end], (start, end), 1]\
                           for start, end in extract.extract_continuous_span(candidates_for_date)]
        
    candidates_for_timedelta = extract.extract_timedelta(text)
    
    candidates_for_amount = extract.extract_continuous_span(extract.extract_amount_spans(text))
    candidates_for_amount =[[text[start:end], (start, end), 1]\
                           for start, end in extract.extract_continuous_span(candidates_for_amount)]
    
    candidates_for_name = extract.extract_name_spans(text)
    
    return dict(date=candidates_for_date, timedelta=candidates_for_timedelta, 
               amount=candidates_for_amount, name=candidates_for_name)
    
```

```python
def normalize_date(dates):
    formated_dt_index = pd.to_datetime(dates, dayfirst=True)
    year = formated_dt_index.year
    month = formated_dt_index.month
    day = formated_dt_index.day
    
    triplets_of_day_month_year = list(zip(day, month, year))
    triplets_of_day_month_year = [list(map(int, i)) if all([float(j).is_integer() for j in i]) else None \
                                   for i in triplets_of_day_month_year]
    return triplets_of_day_month_year
```

```python
def extract_date_spans(text):
    months_and_abbr = '|'.join('january february march april may june july august september october november december'.split())
    months_and_abbr += '|'+'|'.join('jan feb mar apr may jun jul aug sept sep oct nov dec'.split())

    regex_pattern1 = r'(\d{1,4})(.{0,10}?)('+months_and_abbr+')(.{0,5}?)(\d{4})'
    regex_pattern2 = r'(\d{1,4}([.\-/])\d{1,2}([.\-/])\d{1,4})'
    
    span_collection1 = get_index_of(regex_pattern1, [text])
    span_collection2 = get_index_of(regex_pattern2, [text])
    
    span_collection1.extend(span_collection2)
    span_collection1.sort(key=lambda x: x[1][0])
    return span_collection1
```

```python
def get_index_of(regex_pattern=None, subject_strings=None):    
#     print(subject_string, regex_pattern)
    j=0;
    extracted_texts_and_span = []
    for ind, subject_string in enumerate(subject_strings):
#         print(subject_string)
        for i in re.finditer(regex_pattern, subject_string.lower()):
            j+=1
            # print(f'-->in:ind {ind} At:', i.start(), i.end(), 'extracted', i.group())
            extracted_texts_and_span.append([i.group(0), (i.start(), i.end()), ind])
#             extracted_texts_and_span.append(i)
            
    # print(j)
    return extracted_texts_and_span
```
